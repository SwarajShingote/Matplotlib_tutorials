```python
from matplotlib import pyplot as plt

#Adding a Style
plt.style.use('fivethirtyeight')

#Creating a list
#slices=[60,30,10,20]

#Adding labels to slices
#labels=['Sixty','Thirty','Ten','Twenty']

#Adding colors 
#color=['blue','red','yellow','green']

# Note : Generally use hex values for color

# pie() creates pie chart in same proportion of having para as list and labels 
#wedgeprops={'edgecolor' : 'black'} colors edge with black
#plt.pie(slices,labels=labels, colors= color,wedgeprops={'edgecolor' : 'black'} )

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0] # Exploding python by 0.1

#shadow=True adds shadow , startangle= 90  starts chart with 90 degree
#autopct = '%1.1f%%' shows percentage in chart calculated in proportion

plt.pie(slices,labels=labels,explode=explode, shadow = True, 
        startangle = 90, autopct = '%1.1f%%',
        wedgeprops={'edgecolor' : 'black'})



plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
